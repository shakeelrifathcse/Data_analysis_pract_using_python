```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("/content/student_performance.csv")
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
# Boxplot for Scores
plt.figure(figsize=(10, 6))
sns.boxplot(data=data['Score'])
plt.title('Boxplot of Scores')
plt.ylabel('Score')
plt.show()

# Boxplot for Study Hours
plt.figure(figsize=(10, 6))
sns.boxplot(data=data['Study_Hours'])
plt.title('Boxplot of Study Hours')
plt.ylabel('Study Hours')
plt.show()
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x = "Study_Hours", y = "Score" ,hue = "Gender")
plt.title("Study hours vs score")
plt.xlabel("S H")
plt.ylabel("Score")
plt.show()
```

```python
plt.figure(figsize =(10,6))
sns.histplot(data["Score"],bins = 10, kde =True)
plt.title("Distribution of Scores")
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Gender', y='Study_Hours')
plt.title('Study Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Study Hours')
plt.show()
```
